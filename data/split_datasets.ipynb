```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv("fraud_detection_bank_dataset.csv", index_col=0)
df.rename({"targets": "target"}, axis=1, inplace=True)
df.head()
```

```python
X = df.drop("target", axis=1)
y = df["target"]

# X_train to train the model, X_test for hyperparameter tuning and X_val for model validation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.1, stratify=y)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=.2, stratify=y_train)

X_train.to_csv("X_train.csv", index=False)
y_train.to_csv("y_train.csv", index=False)

X_test.to_csv("X_test.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

X_val.to_csv("X_val.csv", index=False)
y_val.to_csv("y_val.csv", index=False)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")
print(f"X_val: {X_val.shape}, y_val: {y_val.shape}")
```
